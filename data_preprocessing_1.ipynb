```python
#Remove unnecessary features
#Remove duplicate names
#Remove nameless games
#combine developers and publishers to dev
#combine categories, genres and tags to full_tags
#Keep only games on windows with english support
#keep only the year from release date
#Remove games with no reviews
#Sort by name
#Re index
#Create final_game_detaills.csv
```

```python
import pandas as pd
games = pd.read_csv("games.csv")
```

```python
print(games)
```

```python
games = games.drop_duplicates(subset=['Name'])
print(games)
```

```python
print(games.columns)
```

```python
shortened_games = games.drop(columns= ['AppID', 'Estimated owners', 'Peak CCU', 'Required age', 'DLC count', 'Price', 'About the game', 'Full audio languages', 'Reviews','Header image', 'Website', 'Support url', 'Support email', 'Metacritic score', 'Mac', 'Linux', 'Metacritic url', 'User score', 'Score rank', 'Achievements', 'Notes', 'Average playtime two weeks', 'Median playtime two weeks', 'Screenshots', 'Movies', 'Supported languages'])
print(shortened_games)
```

```python
shortened_games = shortened_games.dropna(subset=['Name'])
print(shortened_games)
```

```python
# Combine dev and publishers
def combine_and_deduplicate(row, col1, col2):
    # Handle NaN values
    dev = row[col1] if pd.notna(row[col1]) else ''
    pub = row[col2] if pd.notna(row[col2]) else ''
    
    combined = f"{dev},{pub}"
    unique_elements = set(filter(None, combined.split(',')))  # Filter out empty strings
    return ','.join(unique_elements)

# Apply the function to combine 'Developers' and 'Publishers' into 'Dev'
shortened_games['Dev'] = shortened_games.apply(combine_and_deduplicate, col1='Developers', col2='Publishers', axis=1)

# Drop rows where 'Dev' is NaN or empty
shortened_games = shortened_games[shortened_games['Dev'] != '']
shortened_games = shortened_games.dropna(subset=['Dev'])

print(shortened_games)
```

```python
print(shortened_games[shortened_games['Dev'].isna()])
```

```python
shortened_games = shortened_games[shortened_games['Windows']]
print(shortened_games)
```

```python
#Combine Categories , Genres and Tags
def combine_and_deduplicate_tags(row):
    columns_to_combine = ['Categories', 'Genres', 'Tags']
    combined = ','.join(row[col] for col in columns_to_combine if pd.notna(row[col]))
    unique_elements = set(combined.split(','))
    return ','.join(unique_elements)

shortened_games['full_tags'] = shortened_games.apply(combine_and_deduplicate_tags, axis=1)
shortened_games = shortened_games[shortened_games['full_tags'] != '']
shortened_games = shortened_games.dropna(subset=['full_tags'])
print(shortened_games)
```

```python
print(shortened_games[shortened_games['full_tags'].isna()])
```

```python
shortened_games['year'] = shortened_games['Release date'].str[-4:].astype(int)
print(shortened_games)
```

```python
shortened_games = shortened_games[(shortened_games['Positive'] + shortened_games['Negative']) != 0]
```

```python
final_games_details = shortened_games.drop(columns= ['Release date', 'Windows', 'Developers', 'Publishers', 'Categories', 'Genres', 'Tags'])
print(final_games_details)
```

```python
final_games_details = final_games_details.sort_values(by='Positive', ascending= False)
final_games_details.reset_index(drop=True, inplace=True)
final_games_details.insert(0, 'gameID', final_games_details.index)
print(final_games_details)
```

```python
print(final_games_details[final_games_details['full_tags'].isna()])
print(final_games_details[final_games_details['Dev'].isna()])
```

```python
final_games_details['rating'] = 100* final_games_details["Positive"] /(final_games_details["Positive"] + final_games_details["Negative"])
final_games_details['rating'] = final_games_details['rating'].round(2)
```

```python
final_games_details.to_csv("final_games_details.csv", index= False)
```
